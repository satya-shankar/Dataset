```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. LOAD THE DATA
df = pd.read_csv('archive (1).zip')

print("--- Data successfully loaded! ---")

# 2. ASSIGN CORRECT COLUMNS FOR THIS MUSIC TRAFFIC DATASET
timestamp_col = 'date'
page_col = 'track'      # We will treat the song track as the "Page" the user visits

# Since there is no User ID, we combine Country + City to act as a unique user proxy
df['user_proxy'] = df['country'].astype(str) + "_" + df['city'].astype(str)
user_col = 'user_proxy'

# Clean up dates and sort
df[timestamp_col] = pd.to_datetime(df[timestamp_col])
df = df.dropna(subset=[user_col, timestamp_col, page_col])
df = df.sort_values(by=[user_col, timestamp_col]).reset_index(drop=True)

# 3. CREATE SESSIONS (Group clicks together by 30-minute inactivity window)
df['time_diff'] = df.groupby(user_col)[timestamp_col].diff()
df['is_new_session'] = (df['time_diff'].isnull()) | (df['time_diff'] > pd.Timedelta(minutes=30))
df['session_id'] = df.groupby(user_col)['is_new_session'].cumsum()
df['unique_session_id'] = df[user_col].astype(str) + "_" + df['session_id'].astype(str)

# 4. CALCULATE METRICS
total_users = df[user_col].nunique()
total_sessions = df['unique_session_id'].nunique()
session_lengths = df.groupby('unique_session_id').size()
bounce_rate = (session_lengths == 1).sum() / total_sessions * 100

# Average duration (Note: since dates in this dataset are by day, duration defaults to 0 if clicked same day)
session_durations = df.groupby('unique_session_id')[timestamp_col].agg(lambda x: (x.max() - x.min()).total_seconds())
avg_session_duration_sec = session_durations.mean()

print("\n=== YOUR METRICS RESULTS ===")
print(f"Total Unique Regions/Users: {total_users:,}")
print(f"Total Traffic Sessions: {total_sessions:,}")
print(f"Overall Bounce Rate: {bounce_rate:.2f}%")
print(f"Average Session Duration: {avg_session_duration_sec:.2f} seconds")

# 5. GENERATE PLOTS
df['page_sequence'] = df.groupby('unique_session_id').cumcount() + 1
landing_pages = df[df['page_sequence'] == 1][page_col].value_counts().head(10)
exit_page_indices = df.groupby('unique_session_id')[timestamp_col].idxmax()
exit_pages = df.loc[exit_page_indices, page_col].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=landing_pages.values, y=landing_pages.index, palette="viridis")
plt.title("Top 10 Entry Landing Tracks (First Song Played)")
plt.xlabel("Visits")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x=exit_pages.values, y=exit_pages.index, palette="magma")
plt.title("Top 10 Exit Tracks (Last Song Played Before Leaving)")
plt.xlabel("Exits")
plt.tight_layout()
plt.show()
```

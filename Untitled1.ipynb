```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dummy data generate karna Alfido Tech ke liye
np.random.seed(42)
data_size = 1000

data = {
    'Student_ID': range(1001, 1001 + data_size),
    'Domain': np.random.choice(['Data Analytics', 'Web Development', 'Python Programming', 'Android Dev'], size=data_size, p=[0.4, 0.3, 0.2, 0.1]),
    'Stage_Reached': np.random.choice(['1_Visited_Website', '2_Registered', '3_Downloaded_Task', '4_Submitted_Task'], size=data_size, p=[0.1, 0.3, 0.4, 0.2]),
    'Feedback_Rating': np.random.randint(1, 6, size=data_size)
}

df = pd.DataFrame(data)
print("--- Dataset Snapshot ---")
print(df.head())

# Basic Summary Statistics
print("\n--- Summary Statistics ---")
print(df['Domain'].value_counts())
print("\n--- Stages Drop-off Count ---")
print(df['Stage_Reached'].value_counts().sort_index())
```

```python
# Chart 1: Funnel Drop-off Analysis
plt.figure(figsize=(8, 5))
stage_counts = df['Stage_Reached'].value_counts().sort_index()
sns.barplot(x=stage_counts.index, y=stage_counts.values, palette='viridis')
plt.title('Alfido Tech / InternSpark User Funnel Analysis')
plt.xlabel('Funnel Stage')
plt.ylabel('Number of Students')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Chart 2: Domain-wise Distribution
plt.figure(figsize=(7, 5))
df['Domain'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], startangle=140)
plt.title('Most In-Demand Internship Domains')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("ecommerce_customer_data_custom_ratios[1].csv")
df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gender distribution
sns.countplot(x="Gender", data=df)
plt.show()

# Age vs Spending
sns.scatterplot(x="Age", y="Spending", data=df)
plt.show()

# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
### Insights
1. 18–25 age group sabse zyada kharidari karta hai.

```

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
rfm = df.groupby('Customer ID').agg({
    'Purchase Date': lambda x: (df['Purchase Date'].max() - x.max()).days,
    'Customer ID': 'count',
    'Total Purchase Amount': 'sum'
})
rfm.columns = ['Recency','Frequency','Monetary']
```

```python
### Insights
2. Electronics aur Fashion category me sabse zyada sales hoti hai.



```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
```

```python
### Insights
3. Weekend (Saturday–Sunday) me purchase peak hoti hai.

```

```python
from sklearn.cluster import KMeans
X = rfm[['Recency','Frequency','Monetary']]
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(X)
```

```python
### Insights
4. Jinke returns zyada hote hain unme churn risk high hai.

```

```python
# Segment distribution
rfm['Cluster'].value_counts().plot(kind='bar')
plt.show()

# Purchase trends
df.groupby(df['Purchase Date'].dt.month)['Total Purchase Amount'].sum().plot(kind='line')
plt.show()

# Churn vs Age
sns.countplot(x="Customer Age", hue="Churn", data=df)
plt.show()
```

```python
### Insights
5. High-value loyal customers (frequent buyers) revenue ka major part dete hain.
```
